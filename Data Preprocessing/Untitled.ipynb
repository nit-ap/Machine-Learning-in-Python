DATA PREPROCESSING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Our dataset
shirt_order = pd.DataFrame({'Name':['Alex', 'Ben', 'Cam', 'Dave', 'Eli', 'Frank', 'Grace', 'Henry', 'Iris', 'Jack'],
                            'Gender':['F', 'M', 'M', 'M', 'F',  'M', 'F', 'M', 'F', 'M'],
                            'Size':['Small', 'Large', 'Medium', 'Small', 'Medium', 'Large', 'Large', 'Large', 'Small', 'Small'],
                            'Color':['Blue', 'Yellow', 'Red', 'Red', 'Yellow', 'Red', 'Blue', 'Yellow', 'Yellow', 'Blue']
                          })
#Visualize dataset
shirt_order
```

```python
#LABEL ENCODING
from sklearn.preprocessing import LabelEncoder
#Create an object for Label Encoder
LE=LabelEncoder()
#Create a new column that will show you the encoded require column
shirt_order['Gender Category']=LE.fit_transform(shirt_order.Gender)
#Check
shirt_order
```

```python
shirt_order.Size.unique()
```

```python
#ORDINAL ENCODING
mapping_dict={'Small':0,'Medium':1,'Large':2}
shirt_order['Size Category']=shirt_order.Size.map(mapping_dict)
#check
shirt_order
```

```python
#ONE HOT ENCODING
#duplicate color column for keeping the orginal values
shirt_order['color_category']=shirt_order.Color
shirt_order_onehotencoding=pd.get_dummies(shirt_order,columns=["color_category"],prefix=["Color"])
#check
shirt_order_onehotencoding
```
